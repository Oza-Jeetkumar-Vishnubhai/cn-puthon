```python
from socket import * 
import ssl
import base64 
import time

msg = input('Enter the msg you want to send: ') 
endmsg = "\r\n.\r\n"

# Choose a mail server (e.g. Google mail server) and call it mailserver 
mailserver = ('smtp.gmail.com', 465)

# Create socket called clientSocket and establish a TCP connection with mail server 
clientSocket = socket(AF_INET, SOCK_STREAM)
clientSocket = ssl.wrap_socket(clientSocket, ssl_version = ssl.PROTOCOL_SSLv23)
clientSocket.connect(mailserver)
recv = clientSocket.recv(1024).decode() 
print(recv)
if recv[:3] != '220':
    print('220 reply not received from server.')

# Send HELO command and print server response. 
clientSocket.send("HELO Alice\r\n".encode()) 
recv1 = clientSocket.recv(1024).decode() 
print(recv1)
if recv1[:3] != '250':
    print('250 reply not received from server.')

# Since the sender will use base64 to encode the information, the sender needs to input the information
username = "dhrumishah02@gmail.com" 
password = "mathsdoctor"
base64_str = ("\x00"+username+"\x00"+password).encode() 
base64_str = base64.b64encode(base64_str)
authmsg = ("AUTH PLAIN ".encode())+base64_str+("\r\n".encode()) 
clientSocket.send(authmsg) 
recvAuth = clientSocket.recv(1024).decode()
print(recvAuth)

# If the user authentication is successful, the server will return status code 235 
if '235' != recvAuth[:3]:
    print('235 reply not received from server.')
    
# Send MAIL FROM command and print server response. 
clientSocket.send("MAIL FROM:<vazabijal@gmail.com>\r\n".encode()) 
recvFrom = clientSocket.recv(1024).decode()
print("After MAIL FROM command: " + recvFrom) 
if '250' != recvFrom[:3]:
    print('250 reply not received from server.')
    
# Send RCPT TO command and print server response. 
clientSocket.send("RCPT TO:<u20cs021@coed.svnit.ac.in>\r\n".encode()) 
recvTo = clientSocket.recv(1024).decode()
print("After RCPT TO command: " + recvTo) 
if '250' != recvTo[:3]:
    print('250 reply not received from server.')

# Send DATA command and print server response.
clientSocket.send("DATA\r\n".encode()) 
recvData = clientSocket.recv(1024).decode() 
print("After DATA command: " + recvData)
if '354' != recvData[:3]:
    print('354 reply not received from server.')
    
# Edit email information and send data 
subject = "Socket Programming!" 
contentType = "text/plain"
message = 'from:vazabijal@gmail.com \r\nto:u20cs021@coed.svnit.ac.in \r\n' 
message += 'subject:' + subject + '\r\nContent-Type:' + contentType + '\t\n\r\n' + msg 
clientSocket.sendall(message.encode())

# With "." end. Request successful return 250 
clientSocket.send(endmsg.encode()) 
recvEnd = clientSocket.recv(1024).decode() 
print(recvEnd)
if '250' != recvEnd[:3]:
    print('250 reply not received from server.')

# Send the "QUIT" command to disconnect from the mail server 
clientSocket.send("QUIT\r\n".encode())
clientSocket.close()
```
